task 5: convert age to numeric and replace invalid values(negative 0r>100) with NaN

```python

df['age'] = pd.to_numeric(df['age'], errors='coerce')
df.loc[(df['age'] < 0) | (df['age'] > 100), 'age'] = pd.NA
df['age']
```

TASK 6: clean salary by removing currency symbols and converting it to numeric


```python
#6. Clean salary by removing currency symbols and converting it to numeric
df['salary'] = df['salary'].replace('[\$,]', '', regex=True)
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
df['salary']
```

task 7: convert join date into datetime format handiling multiple date formates


```python
#7. Convert join_date to datetime format handling multiple date formats
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce', dayfirst=True)
df['join_date']
```

task :8 extract joining year from the cleaned join date

```python

#8.Extract Joining year from the cleaned join_date
df['joining_year'] = df['join_date'].dt.year
df['joining_year']
```

task :9 standarize departement values by trimming spaces and convertye to upper caser
 

```python
#9.Standardize department values by trimming spaces and converting to uppercase
df['department'] = (df['department'].str.strip().str.upper())
df['department']
```

task:10 Normalize city names to a consistent case and spelling

```python
#10.Normalize city names to a consistent case and spelling
df['city']=df['city'].str.strip().str.title()
df['city']
```

task:11 convert performance score into numeric by mapping text values to numbers


```python
#11.Convert performance score into numeric by mapping text values to numbers
performance_mapping = {'Poor': 0,'Excellent':1}
df['performance_score'] = df['performance_score'].map(performance_mapping)
df['performance_score'] = pd.to_numeric(df['performance_score'],errors='coerce')
df['performance_score']
```

task 12: clean experience years by converting text values to numeric and removing negative 

```python
#12.Clean experience_years by converting text values to numeric and removing negatives
df['experience_years'] = pd.to_numeric(df['experience_years'], errors='coerce')
df.loc[df['experience_years'] < 0, 'experience_years'] = pd.NA
df['experience_years']
```

task 13 : convert is active values (yes/no/True/False) into boolean format 


```python
#13.Convert is_active values(True/False) into boolean format
df['is_active']=df['is_active'].replace({'yes':'True','no':'False'})
df['is_active'].astype('bool')
df['is_active']
```

task 14: identify and remove  dupilcates record based on user id

```python
#14.Identify and remove duplicate records based on user_id
df = df.drop_duplicates(subset="user_id")
df
```

task 15  create a clean final data frame contaning only valid standardized records

```python
clean_df = df.dropna().drop_duplicates().reset_index(drop=True)

clean_df.info()
clean_df.head()
```
